### Making a helper functions

```python
def my_plotter(ax, data1, data2, param_dict):
    """A helper function to make a graph"""
    out = ax.plot(data1, data2, **param_dict)
    return out
```

###### let us the call the function

```python
# data arrays
data1, data2, data3, data4 = np.random.randn(4, 100)
```

```python
print(data1)
```

```python
print(data2)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(5, 2.7))
my_plotter(ax1, data1, data2, {'marker': 'x'})
my_plotter(ax2, data3, data4, {'marker': 'o'})
```

```python
my_plotter(ax1, data1, data2, {'marker': 'x'})
```

# Labelling plots

### 1. Axes labels and text

```python
mu, sigma = 115, 15
x = mu + sigma * np.random.randn(10000)
fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')

# let us create a histogram of the respective data
n, bins, patches = ax.hist(x, 50, density=True, facecolor='C0', alpha=0.75)

# now let us start labelling the components of figure
ax.set_xlabel('Legth [cm]', fontsize=14, color='red')
ax.set_ylabel('Probability', fontsize=14, color='red')
ax.set_title('Sales distribution\n (Volvo cars)')
ax.text(75, 0.025, r'$\mu=115, \ \sigma=15$')
ax.axis([55, 175, 0, 0.03])
ax.grid(True)
```

### Using mathematical expressions in text

```python
mu, sigma = 115, 15
x = mu + sigma * np.random.randn(10000)
fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')

# let us create a histogram of the respective data
n, bins, patches = ax.hist(x, 50, density=True, facecolor='C0', alpha=0.75)

# now let us start labelling the components of figure
ax.set_xlabel('Legth [cm]', fontsize=14, color='red')
ax.set_ylabel('Probability', fontsize=14, color='red')
ax.set_title(r'$\sigma_i=15$')
ax.text(75, 0.025, r'$\mu=115, \ \sigma=15$')
ax.axis([55, 175, 0, 0.03])
ax.grid(True)
```

### Plotting dates and strings

###### dates 

```python
fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
dates = np.arange(np.datetime64('2022-10-15'), np.datetime64('2023-03-15'),
                  np.timedelta64(1, 'h'))
data = np.cumsum(np.random.randn(len(dates)))

# draw time series data
ax.plot(dates, data)

# lets locate the major ticks and format them
cdf = mpl.dates.ConciseDateFormatter(ax.xaxis.get_major_locator())
ax.xaxis.set_major_formatter(cdf)
```

### color mapped data

```python
X, Y = np.meshgrid(np.linspace(-3, 3, 128), np.linspace(-3, 3, 128))
Z = (1 - X/2 + X**5 + Y**3) * np.exp(-X**2 - Y**2)

# let us create a plot
fig, axs = plt.subplots(2, 2, layout='constrained')

# subplot1
pc = axs[0, 0].pcolormesh(X, Y, Z, vmin=-1, vmax=1, cmap='RdBu_r')
fig.colorbar(pc, ax=axs[0, 0])
axs[0, 0].set_title('pcolormesh()')


#subplot2
pc = axs[0, 1].contourf(X, Y, Z, levels=np.linspace(-1.25, 1.25, 11))
fig.colorbar(pc, ax=axs[0, 1])
axs[0, 1].set_title('contourf()')

#subplot3
pc = axs[1, 0].imshow(Z**2 * 100, cmap='plasma', norm=mpl.colors.LogNorm(vmin=0.01, vmax=100))
fig.colorbar(pc, ax=axs[1, 0], extend='both')
axs[1, 0].set_title('imshow() with LogNorm()')

#subplot4
pc = axs[1, 1].scatter(data1, data2, c=data3, cmap='RdBu_r')
fig.colorbar(pc, ax=axs[1, 1], extend='both')
axs[1, 1].set_title('scatter()')
```
